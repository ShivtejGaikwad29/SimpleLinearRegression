```python
# importing the required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# importing the datasets
dataset = pd.read_csv('Salary_Data.csv')
# this is the independant variable or features of matrix
X = dataset.iloc[: , :-1].values
# this is the dependant variable matrix. with the help of this column we are try to predict the salary of the employee
y = dataset.iloc[: ,-1].values
```

```python
# splitting the dataset into the training set and testing set
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.25,random_state=0)
# test size is 0.25% means the 25% of the given dataset
```

```python
# training the linear regression model on training set

from sklearn.linear_model import LinearRegression  #linear_model is the function in linear regression used for training the model
regressor = LinearRegression()
regressor.fit(X_train,y_train)
```

```python
# predicting the test results using predict method
y_pred = regressor.predict(X_test)
```

```python
# visualizing the training set results
plt.scatter(X_train,y_train,color ='yellow')
plt.plot(X_train,regressor.predict(X_train),color='red')
plt.title('SALARY VS TRAINING SET')
plt.xlabel('YEARS OF EXPERIENCE')
plt.ylabel('SALARY')
plt.show()
```
